Basic Charts and Visual Encoding

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = {
    'Days' : ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_hours': [2,3,6,3,4.5],
    'Marks' : [50,60,55,70,65],
    'Attendance' : [66, 70, 85, 92, 75],
    'Sleep_hours' : [7,8,6,8,7],
    'Assignments_completed' : [1,3,2,1,3]
}

df = pd.DataFrame(data)
df
```

```python
#1 Line Chart (Trend)

plt.plot(df['Days'], df['Study_hours'], marker='*')
plt.title('Study Hours Over Days')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
```

```python
# Advanced Line chart

plt.figure(figsize=(9,6))
plt.plot(df['Days'], df['Marks'], marker='o', color = 'violet', label='Marks')
plt.plot(df['Days'], df['Attendance'], marker='x',color = 'green', label='Attendance')
plt.title('Marks and Attendance Over Days')
plt.xlabel('Days')
plt.ylabel('Marks/Attendance')
plt.legend()
plt.show()
```

```python
from matplotlib import color_sequences
#2 Bar Chart (Comparison)

plt.bar(df['Days'], df['Marks'], color = 'pink')
plt.title('Marks comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
# Advanced Bar graph

bars = plt.bar(df['Days'], df['Marks'], color='brown')

for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, str(y),
             ha='center', va='bottom')

plt.title('Marks comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y')
plt.show()
```

```python
#3 Histogram

data = {
    'Marks': [55,45,68,79,45,67,82,80,73,53,58,89,75]
}
df = pd.DataFrame(data)

plt.hist(df['Marks'], bins=5, color='purple', edgecolor='black', alpha=0.56)
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
#4 Scatter Plot (Relationship)
import matplotlib.pyplot as plt

plt.scatter(df['Marks'], df['Attendance'], color='blue')
plt.title('Marks vs Attendance')
plt.xlabel('Marks')
plt.ylabel('Attendance')
plt.show()
```

```python
# Conditional Scatter Plot
# Example Category

df['Result'] = ['Pass', 'Fail', 'Pass', 'Pass', 'Fail']
colors = ['red' if result == 'Fail' else 'green' for result in df['Result']]

plt.scatter(df['Marks'], df['Attendance'], c=colors)
plt.title('Marks vs Attendance')
plt.xlabel('Marks')
plt.ylabel('Attendance')
plt.show()
```

```python
# Dataset-2
data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North',
               'South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics',
                 'Food','Clothing','Electronics','Food']
}

df = pd.DataFrame(data)
df
```

```python
#Line Plot
sns.lineplot(x='Day', y='Sales', data=df)
plt.title("Seaborn Line Plot")
plt.show()

```

```python
# Bar Plot

sns.barplot(x='Region', y='Sales',data=df)
plt.title("Seaborn Bar Plot")
plt.show()
```

```python
# Histogram

sns.histplot(df['Sales'], bins=5, kde=True)
plt.title("Seaborn Histogram")
plt.show()
```

```python
# Scatter Plot

sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Seaborn Scatter Plot")
plt.show()
```
